## **3. 데이터 전처리**
먼저 다음에 4가지에 모두 해당하는 기업을 선별하고자 한다.

1. 연매출액이 한 해라도 1000억 이상을 기록한 기업
2. 직원수가 0명이 아닌 기업
3. 연매출액이 모든 해에 플러스 값을 기록한 기업 
4. 3년 이상의 데이터가 있는 기업

```python
# 1번, 연매출액이 1000억 이상인 기업의 ID추출
temp1 = df[df['연매출액']>100000000].iloc[:,0].unique()
temp1
```

```python
# 2번, 직원수가 0명으로 작성된 기업의 ID추출
temp2 = df[df['월별_직원수']==0].iloc[:,0].unique()
temp2
```

```python
# 3번, 연매출액이 마이너스인 기업의 ID추출
temp3 = df[df['연매출액']<0].iloc[:,0].unique() 
temp3
```

```python
# 1번에 해당하는 기업은 포함하고, 2,3번에 해당하는 기업은 제거한다.
df = df[df['회사ID'].isin(temp1)]
df = df[~df['회사ID'].isin(temp2)]
df = df[~df['회사ID'].isin(temp3)]
```

```python
# 4번, 36개월 이상의 데이터가 있는 경우만 걸리게 필터링한다.
company_df = df.groupby('회사ID').filter(lambda x : len(x) >= 36)
company_df
```

```python
# 전처리 후 남은 기업 개수 확인
len(company_df['회사ID'].unique())
```

```python
company_df.sort_values(by=['회사ID','년도','월'], inplace=True)
company_df
```

## **4. 2차 필터링**
각 기업별로 연매출과 직원수를 연단위로 하여 평균값으로 계산하려고 한다.  
그리고 연매출과 직원수를 직전연도와 비교하여 증가율을 구한 후, 다음 2가지에 해당하는 기업을 제외하여 2차로 선별한다.
1. 전년대비 연매출 증가율이 마이너스를 기록한 기업은 제외
2. 연 평균 직원수의 증가율이 마이너스를 기록한 기업은 제외

```python
# 매출액과 직원수를 연도별 평균 데이터로 확인하기 위해 groupby() 메서드를 사용하여 재구성한다.
group_df = company_df.groupby(['회사ID','년도'])['연매출액','월별_직원수'].mean()
group_df.columns = ['연매출액', '연도별_평균직원수']
group_df['연도별_평균직원수'] = round(group_df['연도별_평균직원수'],1)
group_df.reset_index(inplace=True)

group_df
```

```python
# 연매출액과 연평균 직원수가 전년대비 상승한 비율을 넣을 컬럼을 생성한다.
group_df['연매출액_증가율(%)'] = round(group_df[['회사ID','연매출액']].groupby('회사ID').pct_change(),4) * 100
group_df['연직원수_증가율(%)'] = round(group_df[['회사ID','연도별_평균직원수']].groupby('회사ID').pct_change(),4) * 100
group_df = group_df[['회사ID','년도','연매출액','연매출액_증가율(%)','연도별_평균직원수','연직원수_증가율(%)']]
group_df
```

```python
# 각 기업별 최초연도는 증가율이 결측값으로 나오기 때문에 이것을 0으로 바꾼다
rate_year_df = group_df.fillna(0)
rate_year_df
```

```python
# 1번, 연매출액 증가율이 마이너스인 기업의 ID 추출
temp4 = rate_year_df[rate_year_df['연매출액_증가율(%)']<0].iloc[:,0].unique()
temp4
```

```python
# 2번, 연평균 직원수 증가율이 마이너스인 기업의 ID 추출
temp5 = rate_year_df[rate_year_df['연직원수_증가율(%)']<0].iloc[:,0].unique()
temp5
```

```python
# 연매출과 연평균 직원수가 모두 상승인 기업들로만 추려졌다.
increase_df = rate_year_df[~rate_year_df['회사ID'].isin(temp4)]
increase_df = increase_df[~increase_df['회사ID'].isin(temp5)]
increase_df
```

```python
# 기업수는 15개
len(increase_df['회사ID'].unique())
```

## **5. 데이터 시각화 및 10개 기업선정**

선정된 15개의 기업은 모두 고액의 매출액을 기록했으며, 매해 매출액과 직원수가 증가하고 있는 기업들이다.  
따라서 15개중 10개를 선별할 수 있는 새로운 기준이 필요하다.

```python
company_list = increase_df['회사ID'].unique()
company_list
```

**먼저 15개의 기업의 연매출액과 직원수를 비교해보자**

```python
# 15개 기업의 매출액과 직원수를 꺾은선 그래프로 비교한다.
plt.figure(figsize=(10,8))
sns.pointplot(data=increase_df, x='년도', y='연매출액', hue='회사ID')
plt.title('연도별 매출액')


plt.figure(figsize=(10,8))
sns.pointplot(data=increase_df, x='년도', y='연도별_평균직원수', hue='회사ID')
plt.title('연도별 직원수');
```

15개의 기업 중 매출액과 직원수에서 다른 기업보다 월등히 앞선 2개의 기업을 유니콘 기업으로 선택한다.  
**[227414, 227415]**

**두번째로 연매출액 증가율이 폭등한 기업을 찾아보자. (여기서는 30% 상승으로 기준을 잡았다.)**

```python
# 연매출액 증가율이 30% 이상 증가한 연도가 있는 기업을 확인한다
high_list = increase_df[increase_df['연매출액_증가율(%)']>30].iloc[:,0].unique()

high_list
```

총 7개의 기업이 나왔지만 **데이터를 검토 하던 중에 예상치 못한 문제**가 있었다.

```python
# 7개 기업으로만 연도별 데이터로 구성된 데이터 프레임
final_df = increase_df[increase_df['회사ID'].isin(high_list)]

# 마찬가지로 위 7개 기업의 연매출액 증가율을 그래프로 확인해본다.
plt.figure(figsize=(10,8))
sns.pointplot(data=final_df, x='년도', y='연매출액_증가율(%)', hue='회사ID')
plt.title('연매출액 증가율');

```

여기서 420165 기업은 2019년에 급등한 것으로 나타났지만 실제로는 2018년의 자료가 없다.  
즉 2017년->2019년, 2년 동안의 변화율이기 때문에 직전연도 대비 변화율로 판단한다는 위 조건에 타당하지 않고, 또 2018년도에 어느정도 매출을 달성했는지 확인 할 수 없으므로 제외한다.

따라서 위에 나타난 7개 기업 중 420165를 제외한 6개 기업을 유니콘 기업으로 선정한다.  
**[127366, 232757, 294337, 382850, 403470, 469677]**

**마지막으로 나머지 7개의 기업 중 상대적으로 높은 연매출을 달성한 기업을 찾아보자**

```python
# 기업별 데이터를 그래프로 나타내기 위한 함수 생성
def num_graph(x) :
    data = increase_df[increase_df['회사ID'] == x]
    plt.figure(figsize=(10,8))
    sns.pointplot(data=data, x='년도', y='연매출액')
    plt.title(str(x) + ' 연매출액')
```

```python
# 남은 7개 기업을 리스트로 담고 연매출액 그래프를 각각 그려본다
com_list = [126664,127669,294530,382850,403359,420046,420165]
for i in com_list :
    num_graph(i)
```

그래프를 기업별로 그리니 한눈에 비교하기가 어렵다.  
그래서 하나의 그래프에 모두 나오는 그래프로 다시 그려보았다.

```python
finalgraph_df = increase_df[increase_df['회사ID'].isin(com_list)]
plt.figure(figsize=(10,8))
sns.pointplot(data=finalgraph_df, x='년도', y='연매출액', hue='회사ID')
plt.title('7개 기업의 연도별 매출액');
```

위 그래프를 통해 상위 두 기업이 높은 연매출속에서 꾸준한 상승을 이루었음을 확인했으므로 기업을 유니콘 기업으로 선정한다.
